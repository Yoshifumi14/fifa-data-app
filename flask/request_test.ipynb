```python
import requests
import json
```

```python
PORT = "5000"
base_url = "http://localhost:"+PORT
```

### `/nationality`

```python
res = requests.get(base_url+"/nationality")
print(json.dumps(json.loads(res.json()),indent = 2))
```

### `/players`
`/players?nationality=Japan`

```python
params = {"nationality":"Japan"}
res = requests.get(base_url+"/players",params=params )
# res
print(json.dumps(json.loads(res.json()),indent = 2))
```
